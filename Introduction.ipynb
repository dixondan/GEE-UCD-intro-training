    "# A Few Basic Earth Engine Functionalities\n",
    "\n",
    "## Variable Declaration\n",
    "\n",
    "```JavaScript \n",
    "// Declare a variable with the keyword \"var\"\n",
    "var variableName = ee.Thing('variableValue');\n",
    "\n",
    "```\n",
    "A container object (usually in the form `ee.Thing`) is wrapper of a native `JavaScript` object to a Googleâ€™s servers object\n",
    "\n",
    "\n",
    "\n",
    "## Displaying a Variable or It's Metadata\n",
    "\n",
    "\n",
    "```JavaScript \n",
    "// use the print() to display info in the console\n",
    "print(variableName);\n",
    "```\n",
    "The `print` function accepts a number of inputs and displays them or the metadata associated in the console. In Earth Engine, `print` is limited to 5000 elements. Use print to debug your code\n",
    "\n",
    "\n",
    "## Center the Map\n",
    "\n",
    "```JavaScript \n",
    "Map.setCenter(lon, lat, zoomLevel);\n",
    "```\n",
    "Zoom level varies from 0 (no zoom) to 20 (highest zoom level)\n",
    "\n",
    "## Add a Layer to the Map\n",
    "```JavaScript \n",
    "Map.addLayer(variableName, displayParamsDict, displayName);\n",
    "```\n",
    "\n",

    "## Features and FeatureCollections\n",
    "Features are geometries associated with specific properties\n",
    "Feature collections are groups of features.\n",
    "\n",
    "### Functions\n",
    "A function is a set of instructions to perform a specific task:\n",
    "\n",
    "```JavaScript \n",
    "function functionName(Arguments) {\n",
    "  result = ee.Compute(Arguments);\n",
    "  return result\n",
    "}\n",
    "```\n",
    "\n",
    "### Function Call\n",
    "```JavaScript \n",
    "var result = functionName(input);\n",
    "```\n",
    "This is also known as `unknown function` as opposed to `know functions` as they are assigned to a variable (compare with the first function definition)\n",

    "### Mapping a Function Over a Collection\n",
    "```JavaScript \n",
    "var result = ee.Collection.map(functionName);\n",
    "```\n",
    "Mapping a function over a collection applies the function to every element in the collection.\n",
